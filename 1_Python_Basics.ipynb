### Creating a function to generate `200` random integers between `50` to `500` and return the `sum of integer at odd indices` and `sum of integers at even indices` seperately

```python
import numpy as np
from numpy import random

def sum_odd_even(a):

  ### Sum at Odd Positions
  sum_odd = a[0, 1::2].sum()

  ### Sum at Even Positions
  sum_even = a[0, ::2].sum()

  return sum_odd, sum_even

a = np.random.randint(50,500, size=(1,200))
print(a)
odd_sum, even_sum = sum_odd_even(a)

print("Sum of integers at odd indices:", odd_sum)
print("Sum of integers at even indices:", even_sum)
```

#### Writing a function that generates the first `50` numbers in the Fibonacci sequence and returns them as a list when the function is called.

```python
def generate_fibonacci_series(k):

  l = [0,1]
  term_1 = 0
  term_2 = 1

  while(k>0):
    my_next_term = term_1 + term_2
    l.append(my_next_term)
    term_1 = term_2
    term_2 = my_next_term
    k = k-1

  return l

generate_fibonacci_series(50)
```

#### Creating a function that return the `vowels` from each word when a list of words is passed.

Eg: words = ["generative AI", "Artificial Intelligence", "random", "hallucinations", "Christmas"]

The output of the function must be in the below format:

* Word: generative AI
  * Vowels: `a`, `e`, `i`

* Word: Artificial Intelligence
  * Vowels: `a`, `e`, `i`

* Word: Christmas
  * Volwels: `a`, `i`

The output format with The vowels of each word are arranged in order:
`a, e, i, o, u`.

Input: List of words

Output: As in the example stated above.

```python
def id_vowels(word):
    vowels = "aeiou"
    l = [i for i in word.lower() if i in vowels]

    l.sort()

    new_l = []

    for j in vowels:
      if j in l and j not in new_l:
          new_l.append(j)

    vowel_list = [i for i in new_l if i in vowels]

    print("Word:", word)
    print("Vowels:", vowel_list)

words = ["generative AI", "Artificial Intelligence", "random", "hallucinations", "Christmas"]

for i in words:
  id_vowels(i)
```

### Generating a 4x3 matrix, I will write the code to perform the below operations:

* i. Reshape 4x3 matrix into a 6x2 matrix.
* ii. Transpose the matrix the resultant matrix.
* iii. Multiply each element by the column index of the transposed matrix using broadcasting (for 4x3 matrix).

```python
### Reshaping
l = np.random.randint(20,50, size=(4,3))
print(l)
print(l.reshape(6,2))

### Transpose
s = l.transpose()
print(s)

### Multiplying by broadcasting

for i in range(0,s.shape[1]):
    for j in range(0, s.shape[0]):
        l[i][j] = l[i][j] * i

print("After Multiplying by Broadcasting:\n", l)
```

#### Creating two matrices of size 4x4 with random integers between 1 and 10. The following are performed:
* i. Compute their element-wise sum and product.
* ii. Find the dot product of the two matrices.
* iii. Calculate the determinant and inverse of the first matrix. (Handle the case where the matrix is singular.)

```python
### Element wise sum
r = np.random.randint(1,10, size=(4,4))
print(r)
t = np.random.randint(1,10, size=(4,4))
print(t)

print(r+t)

## Element wise product
print(r*t)

## Dot Product
print(np.dot(r,t))

## Determinant of r
det_r = np.linalg.det(r)
print(det_r)

## Inverse of r
if np.abs(det_r) < 1e-10:
    print("Determinant = 0, so it is a singular matrix. Hence, we cannot compute inverse!")
else:
    inverse_r = np.linalg.inv(r)
    print(inverse_r)
```
